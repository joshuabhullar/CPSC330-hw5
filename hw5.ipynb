````python
results = pd.DataFrame()
````

````python
# From Lecture 5
def mean_std_cross_val_scores(model, X_train, y_train, **kwargs):
    """
    Returns mean and std of cross validation

    Parameters
    ----------
    model :
        scikit-learn model
    X_train : numpy array or pandas DataFrame
        X in the training data
    y_train :
        y in the training data

    Returns
    ----------
        pandas Series with mean scores from cross_validation
    """

    scores = cross_validate(model, X_train, y_train, **kwargs)

    mean_scores = pd.DataFrame(scores).mean()
    std_scores = pd.DataFrame(scores).std()
    out_col = []

    for i in range(len(mean_scores)):
        out_col.append((f"%0.3f (+/- %0.3f)" % (mean_scores.iloc[i], std_scores.iloc[i])))

    return pd.Series(data=out_col, index=mean_scores.index)
````

````python
dummy = DummyClassifier(strategy="stratified")
# cv = pd.DataFrame(cross_validate(dummy, X_train, y_train, return_train_score=True, scoring="f1_macro"))
# summary = cv.agg(["mean", "std"]).T
# summary
results["baseline"] = mean_std_cross_val_scores(dummy, X_train, y_train, return_train_score=True, scoring="f1_macro")
results
````

````python
import shap
from sklearn.metrics import (
    classification_report,
    confusion_matrix,
    ConfusionMatrixDisplay,
    f1_score,
)

# features from Q4
X_train_ready = X_train
X_test_ready = X_test

final_model = make_pipeline(
    preprocessor,
    LGBMClassifier(
        scale_pos_weight=imb_ratio,
        n_estimators=13,
        learning_rate=0.1,
        subsample=0.5,
        random_state=123,
        verbose=-1,
    ),
)

final_model.fit(X_train_ready, y_train)
````

````python
final_preprocessor = final_model.named_steps["columntransformer"]
final_lgbm = final_model.named_steps["lgbmclassifier"]
feature_names = final_preprocessor.get_feature_names_out()

X_train_enc = pd.DataFrame(
    final_preprocessor.transform(X_train_ready),
    columns=feature_names,
    index=X_train_ready.index,
)

X_test_enc = pd.DataFrame(
    final_preprocessor.transform(X_test_ready),
    columns=feature_names,
    index=X_test_ready.index,
)

X_train_enc.head()
````

````python
def get_class1_shap(values):
    if isinstance(values, list):
        return values[1] if len(values) > 1 else values[0]
    if isinstance(values, np.ndarray) and values.ndim == 3:
        return values[:, :, 1]
    return values

explainer = shap.TreeExplainer(final_lgbm)
shap_raw = explainer.shap_values(X_train_enc)
train_shap_values = get_class1_shap(shap_raw)

# base value for class 1
if isinstance(explainer.expected_value, (list, np.ndarray)):
    base_value = explainer.expected_value[1]
else:
    base_value = explainer.expected_value

train_shap_values.shape, base_value
````

````python
importance_df = pd.DataFrame(
    {
        "feature": feature_names,
        "mean_abs_shap": np.abs(train_shap_values).mean(axis=0),
    }
).sort_values("mean_abs_shap", ascending=False).reset_index(drop=True)

importance_df.head(12)
````

````python
shap.summary_plot(train_shap_values, X_train_enc, max_display=15)
````

````python
top_feature = importance_df.loc[0, "feature"]
shap.dependence_plot(top_feature, train_shap_values, X_train_enc)
````

````python
importance_df[importance_df["feature"].str.contains("PAY_|PAY_AMT")].head(10)
````

````python
importance_df[
    importance_df["feature"].str.contains("UTILIZATION|BILL_AMT|LIMIT_BAL")
].head(10)
````

````python
importance_by_group = importance_df.assign(
    group=importance_df["feature"].str.split("__").str[0]
).groupby("group", as_index=True)["mean_abs_shap"].sum().sort_values(ascending=False)

importance_by_group
````

````python
importance_df.head(10)
````

I used SHAP values to interpret my tuned LGBM model. Looking at the global feature importance, the repayment status features (PAY_0, PAY_2, etc) are by far the most important which makes sense. The bill amounts, payment amounts, and the utilization features I created also show up as relevant. On the other hand, features like education and sex don't seem to have much influence on the predictions. For education this might be because of the noisy/unknown values in that column.

From the dependence plot, higher values of PAY_0 (meaning worse repayment history) push the model towards predicting default. This makes sense, as people who are behind on payments are more likely to default.
